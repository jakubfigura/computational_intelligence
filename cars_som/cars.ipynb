### Mamy jak widać dane, gdzie trzeba je zamienić ze stringów na inty, aby model zrozumiał te dane, dlatego że one są porządkowe (tj. zachowują kolejność) korzystamy z tego i stopniujemy je zamieniając odpowiednie wartości kategoryczne na liczbowe

```python
from sklearn.preprocessing import OrdinalEncoder

data_to_encode = {
    "buying":   ["vhigh", "high", "med", "low"],
    "maint":    ["vhigh", "high", "med", "low"],
    "doors":    ["2", "3", "4", "5more"],
    "persons":  ["2", "4", "more"],
    "lug_boot": ["small", "med", "big"],
    "safety":   ["low", "med", "high"]
}
encoder = OrdinalEncoder(categories=[[data_to_encode["buying"],data_to_encode["maint"],data_to_encode["doors"],data_to_encode["persons"],data_to_encode["lug_boot"],data_to_encode["safety"]]])
print(cars_df)
df_encoded = encoder.fit_transform()
```
